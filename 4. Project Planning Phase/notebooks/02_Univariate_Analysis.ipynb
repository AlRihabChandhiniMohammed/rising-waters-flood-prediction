```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("../../3. Project Design Phase/dataset/raw/flood dataset.xlsx")
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

for column in numerical_columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution Plot of {column}')
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

### Observation

The distribution plots indicate that temperature and humidity values are concentrated within a narrow range, showing relatively low variation across the dataset. In contrast, rainfall-related features such as **ANNUAL**, **Jun-Sep**, and **sub** have a wider spread, indicating greater variability in rainfall measurements. The target variable **flood** is highly imbalanced, with non-flood instances occurring more frequently than flood instances.

```python
for column in numerical_columns:
    plt.figure(figsize=(8,2))
    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

## Observation

### **Temperature (Temp)**
- The distribution is very compact with values ranging from **28 to 31**.
- The first quartile (Q1) is **29**, the median is **30**, and the third quartile (Q3) is **31**.
- No outliers are observed.

### **Humidity**
