```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("../../3. Project Design Phase/dataset/raw/flood dataset.xlsx")
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="summer",
    linewidths=1,
    linecolor="black",
    square=True,
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

## Observation

- The heat map illustrates the correlation between all numerical features in the dataset.
- **ANNUAL** and **Jun-Sep** exhibit a **strong positive correlation**, indicating that annual rainfall is largely influenced by rainfall during the June–September season.
- The **flood** feature shows a relatively strong positive correlation with **Jun-Sep** rainfall, suggesting an association between higher monsoon rainfall and flood occurrence.
- Most weather-related features such as **Temperature**, **Humidity**, and **Cloud Cover** have weak correlations with the remaining variables.
- The correlation matrix indicates that only a few feature pairs have strong relationships, while the majority exhibit low to moderate correlation.
